```python
def plot_categorical_column(df, column_name):
    """
    Vẽ biểu đồ tròn và biểu đồ cột cho một cột phân loại trong DataFrame.
    
    Parameters:
    - df: DataFrame chứa dữ liệu
    - column_name: tên cột cần trực quan hóa (kiểu chuỗi)
    """
    data_counts = df[column_name].value_counts()

    plt.figure(figsize=(15, 4))

    # Pie chart
    plt.subplot(1, 2, 1)
    plt.pie(
        data_counts.values,
        labels=data_counts.index,
        autopct='%.2f%%',
        startangle=90,
        labeldistance=1.05
    )
    # plt.legend(loc=4)
    plt.title(f"Tỷ lệ các giá trị trong cột '{column_name}'")

    # Bar chart
    plt.subplot(1, 2, 2)
    bars = plt.bar(
        data_counts.index,
        data_counts.values,
    )

    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            height,
            f'{int(height)}',
            ha='center',
            va='bottom',
            fontsize=10,
            color='red'
        )

    plt.grid(ls='--')
    plt.ylabel('Số lượng')
    plt.xlabel(column_name)
    plt.title(f"Số lượng từng giá trị trong cột '{column_name}'")

    plt.tight_layout()
    plt.show()
```

## Phân tích và kết luận các thuộc tính ảnh hưởng đến tỉ lệ rời bỏ của khách hàng 

```python
# Chuyển đổi {'Yes': 1, 'No': 0} => tính tỉ lệ trung bình khách hàng rời bỏ ở từng thuộc tính trong cột
churn_df['Churn'] = churn_df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
def rate_churn(x):
    colors = {1: 'red', 0: 'blue'}
    ax= sns.countplot(data= churn_df, x= x, hue= 'Churn', palette=colors)
    for container in ax.containers:
        ax.bar_label(container, label_type='edge', fontsize=10)
    plt.title(f'Đồ thị biểu diễn rời bỏ khách hàng dựa trên thuộc tính {x}')
    plt.show()
```

## biểu diễn tỉ lệ rời bỏ các cột dạng số : tenure, MonthlyCharges, TotalCharges

### 17) Thuộc tính tenure 

```python
churn_df['tenure'].sort_values()
```

```python
g = sns.FacetGrid(churn_df, col='Churn')
g.map(plt.hist, 'tenure', bins=40)
plt.show()
```

```python
sns.kdeplot(data=churn_df, x="tenure", hue="Churn", fill=True, common_norm=False, alpha=0.5)
plt.title("Biểu đồ KDE của tenure theo Churn")
plt.xlabel("tenure")
plt.ylabel("Tỉ trọng")
plt.grid(True)
plt.show()
```

**Giải thích**
- fill=True: tô màu dưới đường cong.
- common_norm=False: mỗi nhóm có diện tích riêng biệt, không chuẩn hóa chung (giúp dễ so sánh).
- cột ylable: ở khoảng giá trị đó, xác suất xuất hiện của dữ liệu là bao nhiêu, VÍ DỤ:  Nếu một đỉnh của đường KDE ở tenure/MonthlyCharges/TotalCharges = 70 có giá trị Y = 0.03, điều đó có nghĩa:
Xác suất để một khách hàng có chi phí hàng tháng gần 70 là khoảng 3%.
